```python
import cv2
from google.colab.patches import cv2_imshow # for image display
from google.colab import drive
from skimage import io
drive.mount('/content/gdrive')
```

```python

img = io.imread("/content/gdrive/MyDrive/1009.jpg")
img1 = io.imread("/content/gdrive/MyDrive/one.jpg")
OG_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
OG_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
#cv2_imshow(OG_img)

#Convert image to gray just cause I can
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2_imshow(gray_image1)
```

```python
#orb feature detection
orb = cv2.ORB_create()
test_keypoints, test_descriptor = orb.detectAndCompute(gray_image, None)
test_keypoints1, test_descriptor1 = orb.detectAndCompute(gray_image1, None)
keypoints = cv2.drawKeypoints(gray_image, test_keypoints,None, color = (0, 255, 0))
keypoints1 = cv2.drawKeypoints(gray_image1, test_keypoints1,None, color = (0, 255, 0))
cv2_imshow(keypoints)
cv2_imshow(keypoints1)

# it turns out that it can barely get the feature points of these images
```

```python
#print the position of the feature points
for i in range (0,len(test_keypoints)):
  pts = test_keypoints[i].pt
  print(pts)
```
